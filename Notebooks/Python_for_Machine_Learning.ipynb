## 1.4 Broadcasting

Broadcasting is a powerful feature in NumPy that allows for arithmetic operations between arrays of different shapes by automatically expanding the smaller array to match the dimensions of the larger one. This facilitates flexible and efficient computations without the need for explicit replication of data.

```python
# Broadcasting example
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("\nOriginal Array:\n", arr)
row_vector = np.array([10, 20, 30])
print("\nRow Vector:\n", row_vector)
result = arr + row_vector
print("\nBroadcasting Result:\n", result)
```

The row_vector `[10, 20, 30]` is automatically broadcasted to match the shape of arr `[[1, 2, 3], [4, 5, 6]]`.
The operation is performed element-wise, adding the corresponding elements of both arrays.

## 1.5 Linear Algebra with NumPy
Linear algebra operations are fundamental in many machine learning algorithms, including regression, principal component analysis (PCA), and neural networks. NumPy provides a robust set of linear algebra functions to perform these operations efficiently.

```python
# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
print("\nMatrix A:\n", A)
B = np.array([[5, 6], [7, 8]])
print("\nMatrix B:\n", B)
product = np.dot(A, B)   # np.matmut
print("\nMatrix Multiplication (A dot B):\n", product)
```

```python
# Transpose of a matrix
transpose = A.T
print("\nTranspose of A:\n", transpose)

```

```python
# Inversion of a matrix (if invertible)
C = np.array([[4, 7], [2, 6]])
print("\nMatrix C:\n", C)
inverse_C = np.linalg.inv(C)
print("\nInverse of C:\n", inverse_C)

```

```python
# Solving linear equations
# Solve for x in Ax = b
b = np.array([1, 2])
x = np.linalg.solve(A, b)
print("\nSolution of Ax = b:\n", x)
```

- Matrix Multiplication (np.dot): Multiplies two matrices A and B, resulting in a new matrix where each element is the dot product of the corresponding row and column.

- Transpose (A.T): Flips the matrix over its diagonal, turning rows into columns and vice versa.

- Matrix Inversion (np.linalg.inv): Calculates the inverse of a matrix, if it exists. The inverse is crucial for solving systems of linear equations.

- Solving Linear Equations (np.linalg.solve): Finds the values of x that satisfy the equation Ax = b.

## Fun Exercise 1: NumPy Adventure

### Scenario:

Imagine you're working as a data analyst for a retail company named RetailHub. The company wants to analyze its quarterly sales data for five different products over the course of four quarters to make informed decisions about inventory management, marketing strategies, and sales forecasting. Your task is to create a NumPy array to represent this data and perform several analyses to derive meaningful insights.

```python
import numpy as np

# 1. Create the Sales Data Array
# --------------------------------
# Organizing sales data into a 2D NumPy array where each row represents a product
# and each column represents a quarter.
sales_data = np.array([
    [150, 200, 250, 300],  # Product P1
    [180, 210, 270, 330],  # Product P2
    [120, 160, 200, 240],  # Product P3
    [140, 190, 230, 310],  # Product P4
    [160, 220, 260, 340]   # Product P5
])

print("Sales Data Matrix (Products x Quarters):\n", sales_data)

# 2. Calculate Total Annual Sales per Product
# -------------------------------------------
# Summing sales across all quarters (columns) for each product (row)
total_sales_per_product = np.sum(sales_data, axis=1)
print("\nTotal Annual Sales per Product (in units):\n", total_sales_per_product)

# 3. Calculate Average Quarterly Sales
# -------------------------------------
# Calculating the average sales for each quarter across all products
average_sales_per_quarter = np.mean(sales_data, axis=0)
print("\nAverage Quarterly Sales Across All Products (in units):\n", average_sales_per_quarter)

# 4. Increase Sales by 10%
# -------------------------
# Applying a 10% increase to all sales figures using broadcasting
increased_sales = sales_data * 1.10  # 10% increase
print("\nSales Data After 10% Increase (in units):\n", increased_sales)

# 5. Extract Specific Sales Data
# --------------------------------
# Slicing the array to get sales for the first three products and first two quarters
subset_sales = sales_data[:3, :2]  # Rows 0-2, Columns 0-1
print("\nSubset Sales Data (First 3 Products, First 2 Quarters):\n", subset_sales)

# 6. Compute Matrix Product
# -------------------------
# Calculating the matrix product of sales_data and its transpose to find
# the relationship between different products' sales.
matrix_product = np.dot(sales_data, sales_data.T)
print("\nMatrix Product of Sales Data with its Transpose:\n", matrix_product)

# Interpretation: Each element at (i, j) shows the dot product between Product i and Product j's sales.

# 7. Find Maximum Sales Achieved
# --------------------------------
# Identifying the highest sales figure in the entire sales data
max_sales = np.max(sales_data)
print("\nMaximum Sales Achieved by Any Product in Any Quarter:", max_sales, "units")
```

- Identity Matrix: A square matrix with ones on the diagonal and zeros elsewhere, essential for matrix operations and as a multiplicative identity.

- Random Matrix Generation: Useful for simulations, testing algorithms, and initializing machine learning models.

- Determinant Calculation: Determines if a matrix is invertible and is pivotal in understanding matrix properties.

- Eigenvalues and Eigenvectors: Fundamental in dimensionality reduction techniques like PCA and in understanding matrix behavior.

- Matrix Inversion: Critical for solving linear equations and in algorithms that require inverse matrices.

## 2.4 Handling Missing Data

Real-world datasets often contain missing values due to various reasons like data corruption, entry errors, or incomplete data collection.
Handling these missing values is critical to ensure the integrity and accuracy of your analysis and machine learning models.

```python
display(df)
# Introducing missing values
df.loc[4] = ['Keyboard', 50, None]
print("DataFrame with Missing Values:\n")
display(df)

```

```python
# Checking for missing values
print("\nMissing Values:\n", df.isnull())
type(df.isnull())

```

```python
# Filling missing values
df_filled = df.fillna({'Stock': df['Stock'].mean()})
print("\nDataFrame after Filling Missing Values:\n")
df_filled

```

```python
# Dropping rows with missing values
df_dropped = df.dropna()
print("\nDataFrame after Dropping Missing Values:\n")
df_dropped
```

- Introducing Missing Values: Adding a new row with missing stock information to simulate real-world incomplete data scenarios.

- Detecting Missing Values `isnull()`: Identifies the presence of missing data, returning a boolean DataFrame.

- Filling Missing Values `fillna()`: Replaces NaN values with specified values or statistical measures like mean, median, or mode, ensuring data completeness.

- Dropping Missing Values `dropna()`: Removes rows or columns with missing values, useful when the missing data constitutes a small portion of the dataset or when imputation is not feasible.

### Best Practices:

- **Understanding the Reason for Missing Data:** Before deciding to fill or drop missing values, understand why data is missing to choose the most appropriate handling method.

- **Imputation Techniques:** Depending on the data type and distribution, consider using mean, median, mode, or more sophisticated imputation methods like regression or K-Nearest Neighbors (KNN).

- **Avoiding Data Leakage:** Ensure that imputation or any data manipulation is performed within cross-validation folds to prevent data leakage in machine learning workflows.

## 2.5 Grouping and Aggregation

Grouping and aggregation enable you to summarize and analyze data based on categorical variables, revealing patterns and insights that are pivotal for informed decision-making.

```python
# Creating a sample DataFrame
data_emp = {
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Experience': [5, 3, 7, 2, 4, 6],
    'Salary': [60000, 45000, 75000, 50000, 48000, 52000]
}
emp_df = pd.DataFrame(data_emp)
emp_df

```

```python
# Grouping by Department and calculating average salary
avg_salary = emp_df.groupby('Department')['Salary'].mean()
avg_salary

```

```python
# Grouping by Department and counting employees
count_emp = emp_df.groupby('Department')['Employee'].count()
count_emp

```

```python
# Advanced Aggregation: Multiple aggregations
agg_salary_exp = emp_df.groupby('Department').agg({
    'Salary': ['mean', 'max', 'min'],
    'Experience': 'sum'
})
agg_salary_exp
```

- **Grouping:** The `groupby()` function segments the DataFrame into groups based on unique values in the specified column(s).

- **Aggregation:** Functions like `mean()`, `count()`, `agg()` are used to compute summary statistics for each group.

- **Advanced Aggregation:** The `agg()` function allows for multiple aggregations on different columns simultaneously, providing a richer summary.

### Applications in Machine Learning:

- **Feature Engineering:** Create aggregate features that capture statistics like mean, median, or count for categorical groups.

- **Exploratory Data Analysis (EDA):** Understand how different groups interact with the target variable, guiding model selection and feature importance.

- **Data Normalization:** Normalize data within groups to ensure consistent scaling across different categories.

4. **Analyze passenger survival rates based on gender, class, and embarkation port.**
5. **Visualize the distribution of passenger ages and survival rates across different classes.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the Titanic dataset
titanic = sns.load_dataset('titanic')
print("Titanic Dataset Preview:\n", titanic.head())

# 2. Explore the dataset
print("\nDataset Information:")
titanic.info()

print("\nDescriptive Statistics:\n", titanic.describe())

# 3. Clean the data
# Filling missing 'age' values with the median age
titanic['age_filled'] = titanic['age'].fillna(titanic['age'].median())

# Dropping rows where 'embarked' is missing
titanic_cleaned = titanic.dropna(subset=['embarked'])
print("\nDataFrame after Cleaning:\n", titanic_cleaned.head())

# 4. Analyze passenger survival rates based on gender and class
survival_rates = titanic_cleaned.groupby(['sex', 'class'])['survived'].mean().unstack()
display(survival_rates)

# 5. Visualize the distribution of passenger ages
plt.figure(figsize=(8,6))
plt.hist(titanic_cleaned['age_filled'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()

# Visualizing survival rates across different classes
survival_rates.plot(kind='bar', figsize=(8,6))
plt.title('Survival Rates by Gender and Class')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.legend(title='Class')
plt.ylim(0,1)
plt.show()
```

## Confusion Matrix
Visualize classification performance.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Sample true and predicted labels
y_true = [0, 1, 0, 1, 0, 1, 0, 0, 1, 1]
y_pred = [0, 1, 0, 0, 0, 1, 1, 0, 1, 1]

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### ROC Curve
Evaluate binary classifier performance.


```python
from sklearn.metrics import roc_curve, auc

# Sample data
y_true = [0, 0, 1, 1]
y_scores = [0.1, 0.4, 0.35, 0.8]

fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```

```python
# Multiple Subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 4))

# Sine Plot
axes[0].plot(x, np.sin(x), 'b-', label='sin(x)')
axes[0].set_title('Sine')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('sin(x)')
axes[0].legend()
axes[0].grid(True)

# Cosine Plot
axes[1].plot(x, np.cos(x), 'r-', label='cos(x)')
axes[1].set_title('Cosine')
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('cos(x)')
axes[1].legend()
axes[1].grid(True)

# Tangent Plot
y_tan = np.tan(x)
# Limiting y values to avoid extreme values
y_tan = np.where(np.abs(y_tan) > 10, np.nan, y_tan)
axes[2].plot(x, y_tan, 'g-', label='tan(x)')
axes[2].set_title('Tangent')
axes[2].set_xlabel('X-axis')
axes[2].set_ylabel('tan(x)')
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

Apply the concepts learned to visualize a dataset related to machine learning.

Use the Iris dataset for classification.

```python
# load the dataset

from sklearn.datasets import load_iris
import pandas as pd

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
# Visualize Feature Distributions
# Create histograms for each feature.

df.hist(bins=20, figsize=(12,10), color='skyblue', edgecolor='black')
plt.suptitle('Feature Distributions in Iris Dataset')
plt.show()
```

```python
# Scatter Plot: Petal Length vs. Sepal Length
# Differentiate species by color.
colors = ['r', 'g', 'b']
species = data.target_names

for i, color in enumerate(colors):
    subset = df[df['target'] == i]
    plt.scatter(subset['sepal length (cm)'], subset['petal length (cm)'],
                color=color, label=species[i], alpha=0.6)

plt.title('Petal Length vs. Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```

```python
# Confusion Matrix Example
# Train a simple classifier and visualize the confusion matrix.

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Split data
X = df.drop('target', axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Train classifier
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix for Iris Classification')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
